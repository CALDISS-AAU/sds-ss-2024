```python
# Create an empty list to store the summaries
summaries = []

# Loop through the first 10 texts
for i in tqdm(range(10)):
  # Call the LLM
  chat_completion = client.chat.completions.create(
      model="meta-llama/Meta-Llama-3.1-8B-Instruct-Turbo",
      messages=[
          {
              "role": "system",
              "content": "You are a helpful assistant. You only give me the output and no stupid intros, explanations or notes. Thanks!",
          },
          {
              "role": "user",
              "content": "Summarize following in 1 sentence: " + df['text'][i],
          },
      ],
  )

  # Extract the summary and add it to the list
  output = chat_completion.choices[0].message.content
  summaries.append(output)

# Print the summaries
for summary in summaries:
  print(textwrap.fill(summary, width=80))
```

```python
class ExtractScheme(BaseModel):
    title: str = Field(description="Title of the interview or article")
    publication_date: str = Field(description="Date when the interview or article was published")

    main_topic: str = Field(description="Primary focus of the discussion, e.g., a specific AI technology or trend")

    applications: List[str] = Field(description="AI applications or use cases mentioned")
    technologies: List[str] = Field(description="Specific AI technologies, frameworks, or tools discussed")

    research_and_development: str = Field(description="Overview of research or product development")
    challenges: List[str] = Field(description="Key challenges or limitations mentioned in the context of the AI applications or technologies")

    ethical_issues: Optional[List[str]] = Field(description="Ethical or societal concerns related to the AI applications or research")

    notable_quotes: List[str] = Field(description="Key quotes from the interview or article")

    keywords: List[str] = Field(description="Relevant terms or phrases from the interview or article")
```

```python
# Call the LLM with the JSON schema
chat_completion = client.chat.completions.create(
    model="meta-llama/Meta-Llama-3.1-8B-Instruct-Turbo",
    response_format={"type": "json_object", "schema": ExtractScheme.model_json_schema()},
    messages=[
        {
            "role": "system",
            "content": "You are an AI model tasked with extracting structured information from a text related to AI research and technology. Follow the schema provided below to extract the relevant details. You do not invent information that is not in the provided text.",
        },
        {
            "role": "user",
            "content": "Extract case informatin form following.: " + df['text'][0],
        },
    ],
)
```

```python
extracted_output = json.loads(chat_completion.choices[0].message.content)
print(json.dumps(extracted_output, ensure_ascii=False, indent=2))
```

```python
extracts = []

# Loop through the first 10 texts
for i in tqdm(range(1,11)):
  # Call the LLM
  chat_completion = client.chat.completions.create(
    model="meta-llama/Meta-Llama-3.1-8B-Instruct-Turbo",
    response_format={"type": "json_object", "schema": ExtractScheme.model_json_schema()},
    messages=[
        {
            "role": "system",
            "content": "You are an AI model tasked with extracting structured information from a text related to AI research and technology. Follow the schema provided below to extract the relevant details. You do not invent information that is not in the provided text. You output JSON only. Nothing else.",
        },
        {
            "role": "user",
            "content": "Extract case informatin form following.: " + df['text'][0],
        },
                {
            "role": "assistant",
            "content": good_example_extraction_gpt4o,
        },
                {
            "role": "user",
            "content": "Extract case informatin form following.: " + df['text'][i],
        },
    ],
    )
  extracted_output = json.loads(chat_completion.choices[0].message.content)
  extracts.append(extracted_output)
#
```

```python
extracts
```

```python
# Flatten the structure
flattened_data = []

for entry in extracts:
    flattened_entry = {
        'title': entry['title'],
        'publication_date': entry['publication_date'],
        'main_topic': entry['main_topic'],
        'applications': ', '.join(entry['applications']),
        'technologies': ', '.join(entry['technologies']),
        'research_and_development': entry['research_and_development'],
        'challenges': ', '.join(entry['challenges']),
        'ethical_issues': ', '.join(entry['ethical_issues']) if entry['ethical_issues'] else '',
        'notable_quotes': ', '.join(entry['notable_quotes']) if entry['notable_quotes'] else '',
        'keywords': ', '.join(entry['keywords'])
    }
    flattened_data.append(flattened_entry)

# Convert to DataFrame
flattened_data_df = pd.DataFrame(flattened_data)

# Save to Excel
flattened_data_df.to_excel('research_analysis.xlsx', index=False)
```
